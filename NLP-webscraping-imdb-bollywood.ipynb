```python
from bs4 import BeautifulSoup
import requests

import pandas as pd
from pandas import Series, DataFrame
from ipywidgets import FloatProgress

from time import sleep
from IPython.display import display

import re
import pickle
```

```python
#url='http://www.imdb.com/chart/top?ref_=nv_mv_250_6'
url ='https://www.imdb.com/india/top-rated-indian-movies/?ref_=hm_india_tr_rhs_1'
```

```python
# request the url and download the content using beautiful soup
 # response =requests.get(url)
result = requests.get(url)
```

```python
result # if response 200 means request going through
```

```python
# we want the info in text format and so download the content using beautiful soup 
soup =BeautifulSoup(result.text,'html.parser')

# it will contain all the information from <html> to </html>
```

```python
# now fetch the important area, where the relevant information exists

# to extract the movie name ...we can parse through  <table class ="chart full-width"> and <td
# class="titleColumn">


# we will collect all the movie names
```

```python
summary=soup.findAll('div',{'class':'article'})
```

```python
new_list = soup.find_all('div',{'class':'article'})
#new_list = soup.find_all('h3',{'class':'list-item-header'})

```

```python
for x in new_list:
    for y in x.find_all('a'):
        print(y.text)
```

```python
rating_list = soup.find_all('td',{'class':'ratingColumn'})
```

```python
for x in rating_list:
    for y in x.find_all('strong'):
        print(y.text)
        
#for x in rating_list:
#    for y in x.findall('strong')
 #       print(y)
```
